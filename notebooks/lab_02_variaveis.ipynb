```python
print("Bem-vindo à aula de Python!")
print("Hoje vamos aprender a trabalhar com variáveis, lista e dicionários.\n")
```

```python
#Variáveis simples
nome_usuario = "Luiz"
senha_usuario = 123
print("Usuário:", nome_usuario, "| Senha:", senha_usuario, "\n")
```

```python
#Entrada de dados
print("Vamos criar uma persona!\n")
nome_persona = input("Informe o nome da persona: ")
idade_persona = int(input("Qual é a idade?"))
altura_persona = float(input("Informe a altura (em metros): "))

print("\nOlá", nome_persona, "!")
print("Você tem", idade_persona, "anos e", altura_persona, "m de altura.\n")
```

```python
# Operações simples
falta_100 = 100 - idade_persona
print(nome_persona, "faltam", falta_100, "anos para chegar aos 100 anos!")

```

```python
#Tipos de conversão
print("\n0 tipo de 'falta_100' é:", type(falta_100))
falta_100_txt = str(falta_100)
print("Após conversão, o tipo é:", type(falta_100_txt))
```

```python
# Lista
nomes = ["Luiz", "Maria", "João"]
nomes.append(nome_persona)
print("\nLista de nomes:", nomes)         
```

```python
# Dicionário
persona = {
    "nome": nome_persona,
    "idade": idade_persona,
    "altura": altura_persona
}
print("\nInformações da persona;", persona)
persona["profissão"] = "Estudante"
print("Com profissão adicionada:", persona)
```
